```python
# Load the user dataset
users_df = pd.read_csv('/content/users.csv')
```

```python
users_df
```

```python
from surprise import Reader, Dataset, KNNBasic
from surprise.model_selection import train_test_split


# Create a merged dataset with required columns for collaborative filtering
merged_df = pd.merge(users_df[['User-ID', 'Genres', 'Ratings']], books_df_init[['Genres', 'Book-Title']], on='Genres')

# Create a Surprise Reader object
reader = Reader(rating_scale=(1, 10))

# Load the merged dataset into Surprise Dataset format
data = Dataset.load_from_df(merged_df[['User-ID', 'Genres', 'Ratings']], reader)

# Split the dataset into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2)

# Use the KNNBasic algorithm for collaborative filtering
algo = KNNBasic()

# Train the model on the training set
algo.fit(trainset)

# Get recommendations for a specific user
user_id = 123  # Replace with the user ID for whom you want recommendations
user_books = merged_df.loc[merged_df['User-ID'] == user_id, 'Genres'].unique()

# Remove books that the user has already rated
all_books = merged_df['Genres'].unique()
recommended_books = [book for book in all_books if book not in user_books]

# Get the top 5 recommended books for the user
top_recommended_books = []
for book in recommended_books:
    prediction = algo.predict(uid=user_id, iid=book, verbose=False)
    top_recommended_books.append(prediction)

# Sort the recommendations by predicted rating in descending order
top_recommended_books = sorted(top_recommended_books, key=lambda x: x.est, reverse=True)[:5]

# Print the top recommended books
print("Top recommended books for User ID", 'unza')
for recommendation in top_recommended_books:
    book_title = books_df_init.loc[books_df_init['Genres'] == recommendation.iid, 'Book-Title'].values[0]
    print(f"Book: {book_title}, Predicted Rating: {recommendation.est}")
```
